```python
import numpy as np
from functools import reduce

def task1(x):
    size = min(len(x), len(x[0]))
    result = reduce(lambda res, y: res *
                    y, [x[i][i] for i in range(0, size) if x[i][i] != 0], 1)
    return result

task1([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]])
```

```python
def np_task1(x):
    d = x.diagonal()
    d = d[d != 0]
    return np.prod(d)
    
np_task1(np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]]))
```

```python
def task2(x, y):
    return sorted(x) == sorted(y)

task2([1, 2, 2, 4], [4, 2, 1, 2])
```

```python
def np_task2(x, y):
    return np.array_equal(np.sort(x), np.sort(y))

np_task2(np.array([1, 2, 2, 4]), np.array([4, 2, 1, 2]))
```

```python
def task3(x):
    new_arr = [x[i + 1] for i in range(len(x) - 1) if x[i] == 0]
    return max(new_arr)

task3([6, 2, 0, 3, 0, 0, 5, 7, 0])
```

```python
def np_task3(x):
    temp_arr = np.concatenate([[1], x])[:-1]
    arr_next = x[np.where(temp_arr == 0)]
    return max(arr_next)

np_task3(np.array([6, 2, 0, 3, 0, 0, 5, 7, 0]))
```

```python
from scipy.misc import imread
from matplotlib import pyplot as plt
%matplotlib inline

def task4(img, coefs):
    result = []
    for line in img:
        temp_res = []
        for k in line:
            temp_res.append(sum([i * j for i, j in zip(k, coefs)]))
        result.append(temp_res)
    return result
```

```python
def np_task4(img, coefs):
    result = np.dot(img[..., :3], coefs)
    return result
```

```python
vec = [0.299, 0.587, 0.114]
im = imread('cat.png', mode='RGB')
# gray_im = task4(im, vec)
gray_im = np_task4(im, vec)

h, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5)) 
ax1.imshow(im, cmap=plt.cm.gray) 
ax2.imshow(gray_im, cmap=plt.cm.gray) 
ax1.set_title("Before") 
ax2.set_title("After")
```

```python
def task5(x):
    elems, count = [x[0]], [0]
    for i in x:
        if elems[-1] == i:
            count[-1] += 1
        else:
            elems.append(i)
            count.append(1)
    return elems, count

keys, values = task5([2, 2, 2, 3, 3, 3, 5, 2])
print(keys, values)
```

```python
def np_task5(x):
    inds = np.concatenate([np.where(np.diff(x) != 0)[0], [len(x) - 1]])
    values = x[inds]
    count = np.concatenate([[inds[0] + 1], np.diff(inds)])
    return values, count

np_task5(np.array([2, 2, 2, 3, 3, 3, 5]))
```

```python
import math

def task6(x, y):
    result = []
    for l in x:
        st = []
        for m in y:
            st.append(math.sqrt(sum((i - j)**2 for i, j in zip(l, m))))
        result.append(st)
    return result
```

```python
def np_task6(x, y):
    dif = np.sum((x[:, np.newaxis] - y) ** 2, axis=2)  
    return np.sqrt(dif)
```
